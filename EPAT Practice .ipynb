### VRADS - VAC Block 1 Full Analysis Refactored into Functions (Modular Python)
### Time Series, Volatility Analysis, ARIMA, ARIMAX, GARCH, RV/HV with RSI and ATR

```python
# VRADS - VAC Block 1 Full Analysis Refactored into Functions (Modular Python)
# Time Series, Volatility Analysis, ARIMA, ARIMAX, GARCH, RV/HV with RSI and ATR

import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima.model import ARIMA
from arch import arch_model

# -----------------------------
# Function 1: Data Collection
# -----------------------------
def fetch_data(ticker, start, end):
    df = yf.download('HDFCBANK.NS', start='2014-01-01', end='2024-12-31', interval='1d')
    return df[['Open', 'High', 'Low', 'Close', 'Volume']].dropna()

# -----------------------------
# Function 2: Sanity Check and Initial Visualization
# -----------------------------
def sanity_check(df):
    print(df.isnull().sum())
    print(df.describe())
    df['Close'].plot(title='Close Price')
    plt.show()

# -----------------------------
# Function 3: Stationarity Check and Returns Calculation
# -----------------------------
def stationarity_and_returns(df):
    adf_result = adfuller(df['Close'])
    print('ADF Statistic:', adf_result[0])
    print('p-value:', adf_result[1])

    if adf_result[1] > 0.05:
        df['Returns'] = np.log(df['Close'] / df['Close'].shift(1))
    else:
        df['Returns'] = df['Close'].pct_change()

    return df.dropna()

# -----------------------------
# Function 4: RV and HV Calculation
# -----------------------------
def calculate_rv_hv(df, window=20):
    df['RV'] = df['Returns'].rolling(window=window).apply(lambda x: np.sqrt(np.sum(x**2)))
    df['HV'] = df['Returns'].rolling(window=window).std() * np.sqrt(window)
    return df

# -----------------------------
# Function 5: ARIMA Baseline Model
# -----------------------------
def run_arima(df):
    model = ARIMA(df['Close'], order=(1,1,1))
    result = model.fit()
    print(result.summary())
    return result

# -----------------------------
# Function 6: RSI and ATR Calculation
# -----------------------------
def calculate_rsi_atr(df):
    delta = df['Close'].diff()
    # gain = np.where(delta > 0, delta, 0)
    # loss = np.where(delta < 0, -delta, 0)
    # df['avg_gain'] = pd.Series(gain).rolling(window=14).mean()
    # df['avg_loss'] = pd.Series(loss).rolling(window=14).mean()
    
    gain = delta.clip(lower=0)
    loss = -delta.clip(upper=0)

    df['avg_gain'] = gain.rolling(window=14).mean()
    df['avg_loss'] = loss.rolling(window=14).mean()
    rs = df['avg_gain'] / df['avg_loss']
    df['RSI'] = 100 - (100 / (1 + rs))

    df['H-L'] = df['High'] - df['Low']
    df['H-PC'] = abs(df['High'] - df['Close'].shift(1))
    df['L-PC'] = abs(df['Low'] - df['Close'].shift(1))
    df['TR'] = df[['H-L', 'H-PC', 'L-PC']].max(axis=1)
    df['ATR'] = df['TR'].rolling(window=14).mean()

    return df.dropna()

# -----------------------------
# Function 7: ARIMAX Model with Exogenous Variables
# -----------------------------
def run_arimax(df):
    exog_vars = df[['RSI', 'ATR', 'RV', 'HV']]
    model = ARIMA(df['Close'], exog=exog_vars, order=(1,1,1))
    result = model.fit()
    print(result.summary())
    return result

# -----------------------------
# Function 8: ARIMAX Forecast
# -----------------------------
def forecast_arimax(result, df):
    exog_vars = df[['RSI', 'ATR', 'RV', 'HV']]
    forecast = result.forecast(steps=5, exog=exog_vars.tail(5))
    print("ARIMAX Price Forecast:")
    print(forecast)
    return forecast

# -----------------------------
# Function 9: GARCH Volatility Modeling
# -----------------------------
def run_garch(df):
    model = arch_model(df['Returns'] * 100, vol='Garch', p=1, q=1)
    result = model.fit()
    print(result.summary())
    return result

# -----------------------------
# Function 10: GARCH Forecast
# -----------------------------
def forecast_garch(result):
    garch_forecast = result.forecast(horizon=5)
    print("GARCH Volatility Forecast:")
    print(garch_forecast.variance.tail())
    return garch_forecast

# -----------------------------
# Running the VRADS Full Flow Modularly
# -----------------------------

df = fetch_data('HDFCBANK.NS', '2014-01-01', '2024-12-31')
sanity_check(df)
df = stationarity_and_returns(df)
df = calculate_rv_hv(df)
arima_result = run_arima(df)
df = calculate_rsi_atr(df)
arimax_result = run_arimax(df)
forecast_arimax(arimax_result, df)
garch_result = run_garch(df)
forecast_garch(garch_result)

# -----------------------------
# Final Note: This functional breakdown makes the VRADS VAC block scalable,
# maintainable, and easier to debug, test, or modify independently.
# -----------------------------
```

## 6 months dummy data, for VAC

```python
import pandas as pd
import numpy as np
import datetime as dt

# HDFC Data Settings
start_date = dt.datetime(2024, 4, 1)
end_date = dt.datetime(2024, 10, 1)
date_range = pd.date_range(start=start_date, end=end_date, freq='D')

spot_price = 1500  # Initial spot price
strikes = np.arange(1400, 1600, 20)  # Strike prices from 1400 to 1580

# Weekly (Fridays) and Monthly (Month-end Thursdays) Expiries
weekly_expiries = pd.date_range(start=start_date, end=end_date, freq='W-FRI')
monthly_expiries = pd.date_range(start=start_date, end=end_date, freq='M') + pd.offsets.Week(weekday=3)

# Simulate HDFC Volatility Index (Mean-reverting process)
np.random.seed(42)
hdfc_vol_index = 22 + 2 * np.random.randn(len(date_range))  # Avg ~22%

def calculate_iv(strike, atm_strike, atm_iv, market_regime):
    """Generate IV using a quadratic model with market regime adjustments."""
    a = 0.00005  # Controls smile intensity
    b = 0.0001 if market_regime == 'bullish' else -0.0001  # Skew direction
    c = atm_iv  # ATM IV
    iv = a * (strike - atm_strike) ** 2 + b * (strike - atm_strike) + c
    return round(max(iv, 0.07), 4)  # IV floor adjusted to 7%

data = []
for i, (current_date, vol_idx) in enumerate(zip(date_range, hdfc_vol_index)):
    # Simulated mean-reverting spot price
    spot_price *= (1 + np.random.normal(0, 0.005))  # Small drift
    
    # Determine market regime
    if vol_idx > 24:
        market_regime = 'bearish'  # High volatility â†’ Put IV > Call IV
    elif vol_idx < 20:
        market_regime = 'bullish'  # Low volatility â†’ Call IV > Put IV
    else:
        market_regime = 'neutral'
    
    # Define ATM IV (baseline)
    atm_iv = 0.22 + np.random.normal(0, 0.005)
    
    for expiry in sorted(set(weekly_expiries).union(set(monthly_expiries))):
        if expiry <= current_date:
            continue  # Skip expired
        
        time_to_expiry = (expiry - current_date).days / 365
        
        for strike in strikes:
            for option_type in ['Call', 'Put']:
                option_iv = calculate_iv(strike, spot_price, atm_iv, market_regime)
                
                # More realistic pricing (OTM options are cheaper)
                intrinsic_value = max(spot_price - strike, 0) if option_type == 'Call' else max(strike - spot_price, 0)
                option_price = intrinsic_value + np.exp(-0.05 * time_to_expiry) * np.random.uniform(10, 30)
                
                # Correlated Open Interest, OI Change, and Volume
                oi_change = np.random.randint(-200, 200)
                oi = max(3000 + oi_change, 3000)  # Base OI liquidity of 3000
                volume = max(np.random.randint(50, 300) + abs(oi_change // 5), 50)
                
                # Smooth IV for low liquidity options
                if volume < 100:
                    option_iv = (option_iv + atm_iv) / 2  
                
                data.append({
                    'Date': current_date.strftime('%Y-%m-%d'),
                    'Expiry': expiry.strftime('%Y-%m-%d'),
                    'Spot': round(spot_price, 2),
                    'Strike': strike,
                    'Type': option_type,
                    'IV': option_iv,
                    'Option_LTP': round(option_price, 2),
                    'OI': oi,
                    'OI_Change': oi_change,
                    'Volume': volume,
                    'Time_to_Expiry': round(time_to_expiry, 4),
                    'HDFC_Vol_Index': round(vol_idx, 2)
                })

# Convert to DataFrame
df = pd.DataFrame(data)

# Add Days to Expiry
df['Days_to_Expiry'] = (pd.to_datetime(df['Expiry']) - pd.to_datetime(df['Date'])).dt.days

# Export CSV
df.to_csv('hdfc_Dummy_options_data.csv', index=False)

# Preview output

print(df.head(10))
```

## Block 1, VAC Layer

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit
import warnings
warnings.filterwarnings("ignore")

# Load the rich 6-month HDFC dummy options dataset
options_df = pd.read_csv('hdfc_dummy_options_data.csv')

# Ensure proper datetime parsing
options_df['Date'] = pd.to_datetime(options_df['Date'])
options_df['Expiry'] = pd.to_datetime(options_df['Expiry'])

# Filter to find the nearest valid future expiry (excluding expired contracts)
current_date = options_df['Date'].min()  # Use the first date in the dataset as 'today'
future_options = options_df[options_df['Expiry'] > current_date]
nearest_expiry = future_options['Expiry'].min()

# Extract options data for the nearest expiry
near_expiry_data = future_options[future_options['Expiry'] == nearest_expiry].copy()

# Approximate spot price for this expiry (taking mean if fluctuates intraday)
spot_price = near_expiry_data['Spot'].mean()

# Dynamically find the ATM Strike closest to Spot
atm_strike = near_expiry_data.iloc[(near_expiry_data['Strike'] - spot_price).abs().argsort()[:1]]['Strike'].values[0]
atm_iv = near_expiry_data[near_expiry_data['Strike'] == atm_strike]['IV'].values[0]

print(f" Spot Price (approx): {round(spot_price, 2)}, ATM Strike: {atm_strike}, ATM IV: {round(atm_iv, 4)}")

# Separate Calls and Puts for IV Smile analysis
calls = near_expiry_data[near_expiry_data['Type'] == 'Call']
puts = near_expiry_data[near_expiry_data['Type'] == 'Put']

# Define a quadratic function for IV Smile fitting
def iv_smile_fit(strike, a, b, c):
    return a * (strike - atm_strike) ** 2 + b * (strike - atm_strike) + c

# Fit the quadratic IV Smile curve for Calls and Puts
popt_calls, _ = curve_fit(iv_smile_fit, calls['Strike'], calls['IV'])
calls['IV_Fitted'] = iv_smile_fit(calls['Strike'], *popt_calls)

popt_puts, _ = curve_fit(iv_smile_fit, puts['Strike'], puts['IV'])
puts['IV_Fitted'] = iv_smile_fit(puts['Strike'], *popt_puts)

# Compute Put-Call IV Ratio (PCIVR)
pcivr = puts.set_index('Strike')['IV'] / calls.set_index('Strike')['IV']
pcivr = pcivr.fillna(1)  # Default to 1 if missing data

# Compute IV Skew: Difference from ATM IV, normalized by Strike Distance
near_expiry_data['IV_Skew'] = (near_expiry_data['IV'] - atm_iv) / np.abs(near_expiry_data['Strike'] - atm_strike)
near_expiry_data['IV_Skew'] = near_expiry_data['IV_Skew'].replace([np.inf, -np.inf], np.nan).fillna(0)

# Compute Realized Volatility (RV) from Spot prices in the entire dataset
options_df.sort_values(by='Date', inplace=True)
options_df['Returns'] = options_df['Spot'].pct_change()
options_df['RV'] = options_df['Returns'].rolling(window=10).std() * np.sqrt(252)  # 10-day window
options_df['RV'] = options_df['RV'].interpolate(method='linear')
current_rv = options_df['RV'].iloc[-1]  # Most recent RV value

# Plot IV Smile for Calls and Puts with RV overlay
plt.figure(figsize=(10, 5))
plt.scatter(calls['Strike'], calls['IV'], marker='o', label='Call IV')
plt.scatter(puts['Strike'], puts['IV'], marker='s', label='Put IV', alpha=0.7)
plt.plot(calls['Strike'], calls['IV_Fitted'], linestyle='-', color='b', label='Call IV Smile Fit')
plt.plot(puts['Strike'], puts['IV_Fitted'], linestyle='-', color='r', label='Put IV Smile Fit')
plt.axhline(current_rv, linestyle='--', color='gray', label=f'RV ({round(current_rv, 4)})')
plt.axvline(atm_strike, linestyle='--', color='black', label='ATM Strike')
plt.title(f'HDFC IV Smile - Nearest Expiry ({nearest_expiry.date()})')
plt.xlabel('Strike Price')
plt.ylabel('Implied Volatility (IV)')
plt.legend()
plt.grid(True)
plt.show()

# Save outputs for downstream VAC Blocks
near_expiry_data.to_csv('vac_block1_output_iv_skew.csv', index=False)
options_df.to_csv('vac_block1_output_full_with_rv.csv', index=False)
pcivr.to_csv('vac_block1_output_pcivr.csv')


near_expiry_data.head(10)
```

## Block 3, VAC Layer

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from hmmlearn.hmm import GaussianHMM
from sklearn.preprocessing import StandardScaler

# Load IV Skew output from Block 1
block1_iv_skew_df = pd.read_csv('vac_block1_output_iv_skew.csv')
block1_iv_skew_df['Date'] = pd.to_datetime(block1_iv_skew_df['Date'])

# Load the original options data to compute RV and HV
options_df = pd.read_csv('hdfc_dummy_options_data.csv')
options_df['Date'] = pd.to_datetime(options_df['Date'])

# Load Term Structure from Block 2 (IV Curve across Expiries)
term_structure = options_df.groupby('Date')['IV'].mean().reset_index().rename(columns={'IV': 'Avg_IV_Term'})

# Calculate Realized Volatility (RV) based on Spot prices
options_df = options_df.sort_values(by='Date')
options_df['Spot_Return'] = np.log(options_df['Spot'] / options_df['Spot'].shift(1))
options_df['RV'] = options_df['Spot_Return'].rolling(window=5).std() * np.sqrt(252)

# Historical Volatility (HV) as long-term rolling standard deviation of Spot returns
options_df['HV'] = options_df['Spot_Return'].rolling(window=20).std() * np.sqrt(252)

# Aggregate IV Skew at Date level (mean skew per day)
daily_skew = block1_iv_skew_df.groupby('Date')['IV_Skew'].mean().reset_index()

# Merge all inputs: RV, HV, IV Skew, Term Structure
merged = pd.merge(options_df[['Date', 'RV', 'HV']].drop_duplicates(), daily_skew, on='Date', how='inner')
merged = pd.merge(merged, term_structure, on='Date', how='inner')

# Drop any NaN rows due to rolling computations
merged = merged.dropna().reset_index(drop=True)

# Normalize the data for HMM input
features = merged[['RV', 'HV', 'IV_Skew', 'Avg_IV_Term']]
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Hidden Markov Model for Regime Detection
hmm_model = GaussianHMM(n_components=4, covariance_type='full', n_iter=100, random_state=42)
hmm_model.fit(features_scaled)

# Predict Regimes
merged['Regime'] = hmm_model.predict(features_scaled)

# Map regimes to descriptive labels (Optional: tweak mapping based on inspection)
regime_map = {0: 'Low Vol', 1: 'Normal', 2: 'High Vol', 3: 'Crisis'}
merged['Regime_Label'] = merged['Regime'].map(regime_map)

# Output Sample
print("\n Block 3 - Regime Detection Sample Output:\n", merged[['Date', 'RV', 'HV', 'IV_Skew', 'Avg_IV_Term', 'Regime_Label']].head())

# Plot Regime Shifts
plt.figure(figsize=(14, 6))
colors = {'Low Vol': 'green', 'Normal': 'blue', 'High Vol': 'orange', 'Crisis': 'red'}
for label, group in merged.groupby('Regime_Label'):
    plt.scatter(group['Date'], group['RV'], label=label, color=colors[label], s=20)

plt.title('VRADS Block 3 - Regime Detection (RV Colored by Regime)')
plt.xlabel('Date')
plt.ylabel('Realized Volatility (RV)')
plt.legend()
plt.grid(True)
plt.show()

# Save the regime-labeled dataset for downstream strategies
merged.to_csv('vac_block3_regime_output.csv', index=False)
```

## Block 4, VAC Layer

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Load Block 3 Regime Detection Output (Regimes already computed)
regime_df = pd.read_csv('vac_block3_regime_output.csv')
regime_df['Date'] = pd.to_datetime(regime_df['Date'])

# Load Dummy Options Data enriched with IV_Skew, OI_Change, and Spot for RV calculation
options_df = pd.read_csv('vac_block1_output_iv_skew.csv')  # Pull IV_Skew from Block 1 output
options_df['Date'] = pd.to_datetime(options_df['Date'])

# Merge regime labels from Block 3 with options data (containing IV_Skew)
merged = pd.merge(options_df, regime_df[['Date', 'Regime_Label']], on='Date', how='inner')

# Calculate Realized Volatility (RV) using rolling returns (5-day window as example)
merged['Returns'] = np.log(merged['Spot'] / merged['Spot'].shift(1))
merged['RV'] = merged['Returns'].rolling(window=5).std() * np.sqrt(252)  # Annualized RV

# Generate Dummy NIFTY VIX (India VIX) Data - Centered around 17 with random stress spikes
np.random.seed(42)
merged['NIFTY_VIX'] = np.random.normal(loc=17, scale=2, size=len(merged))
merged['NIFTY_VIX'] = np.clip(merged['NIFTY_VIX'], 14, 25)

# Random VIX spikes to simulate stress events (10% data)
spike_indices = merged.sample(frac=0.1, random_state=42).index
merged.loc[spike_indices, 'NIFTY_VIX'] += np.random.uniform(5, 10, size=len(spike_indices))

# Market Stress Score Composite - Now includes RV, along with NIFTY_VIX, OI_Change, IV_Skew
scaler = StandardScaler()
stress_features = merged[['NIFTY_VIX', 'OI_Change', 'IV_Skew', 'RV']].fillna(0)  # Ensure no NaNs
stress_scaled = scaler.fit_transform(stress_features)

# Weighted Stress Score - You can tweak weights as per research preference
merged['Stress_Score'] = (
    0.35 * stress_scaled[:, 0] +  # NIFTY_VIX
    0.30 * stress_scaled[:, 1] +  # OI_Change
    0.20 * stress_scaled[:, 2] +  # IV_Skew
    0.15 * stress_scaled[:, 3]    # Realized Volatility (RV)
)

# Quantile-based Stress Classification - Top 20% triggers early crash warning
merged = merged.dropna(subset=['Stress_Score']).reset_index(drop=True)
merged['Stress_Quantile'] = pd.qcut(merged['Stress_Score'], q=5, labels=[1, 2, 3, 4, 5])

# Early Warning Signal - If Stress Quantile is in top 20% (4 or 5)
merged['Crash_Warning'] = np.where(merged['Stress_Quantile'].astype(int) >= 4, 1, 0)

# Plot Market Stress Score with Crash Warning Zones
plt.figure(figsize=(14, 6))
plt.plot(merged['Date'], merged['Stress_Score'], label='Market Stress Score', color='purple')
plt.fill_between(merged['Date'], merged['Stress_Score'], where=merged['Crash_Warning'] == 1,
                 color='red', alpha=0.3, label='Crash Warning Zone')
plt.title('VRADS Block 4 - Market Stress Score with Crash Warning Zones (Includes IV_Skew & RV)')
plt.xlabel('Date')
plt.ylabel('Stress Score')
plt.legend()
plt.grid(True)
plt.show()

# Optional: Print potential early crash warning dates
warnings = merged[merged['Crash_Warning'] == 1][['Date', 'Stress_Score', 'Regime_Label', 'RV']]
print("\n Potential Crash Warning Dates:\n", warnings.head(10))

# Save Block 4 output for Block 5 or Research - Ensure RV is included
merged.to_csv('vac_block4_crash_prediction_output.csv', index=False)


```

## Block 5, VAC Layer

```python
import numpy as np
import pandas as pd
from statsmodels.tsa.api import VAR

# Step 1: Load the enriched Block 4 output (includes IV_Skew, RV, Stress_Score, Regime_Label, etc.)
block5_df = pd.read_csv('vac_block4_crash_prediction_output.csv')
block5_df['Date'] = pd.to_datetime(block5_df['Date'])

# Step 2: Print columns once to verify what's inside
print("Columns Available:", block5_df.columns.tolist())

# Step 3: Handle possible RV naming difference
rv_column = 'RV' if 'RV' in block5_df.columns else 'Realized_Volatility' if 'Realized_Volatility' in block5_df.columns else None

if rv_column is None:
    raise KeyError(" 'RV' or 'Realized_Volatility' column not found in dataset!")

# Step 4: Ensure numeric types for modeling
for col in ['IV', 'IV_Skew', rv_column, 'Stress_Score', 'Regime_Label']:
    block5_df[col] = pd.to_numeric(block5_df[col], errors='coerce')

# Step 5: Prepare VAR Model Dataset - Focus on Volatility Metrics
var_cols = ['IV', 'IV_Skew', rv_column]
var_data = block5_df[var_cols].dropna()

# Optional: Remove columns with negligible variance
var_data = var_data.loc[:, var_data.var() > 1e-5]
print(f" VAR dataset ready | Shape: {var_data.shape} | Columns: {list(var_data.columns)}")

# Step 6: Run VAR Model - Multi-dimensional Volatility Forecast
if var_data.shape[0] >= 10 and var_data.shape[1] >= 2:
    try:
        var_model = VAR(var_data)
        var_result = var_model.fit(maxlags=1, ic='aic')
        var_forecast = var_result.forecast(var_data.values[-1:], steps=1)

        # Extract forecasts for IV_Skew and RV
        forecast_iv_skew = var_forecast[0][var_data.columns.get_loc('IV_Skew')] if 'IV_Skew' in var_data.columns else np.nan
        forecast_rv = var_forecast[0][var_data.columns.get_loc(rv_column)] if rv_column in var_data.columns else np.nan

        print(f" VAR Forecast -> IV_Skew: {forecast_iv_skew:.4f}, RV: {forecast_rv:.4f}")
    except Exception as e:
        print(f" VAR Model Failure: {e}")
        forecast_iv_skew, forecast_rv = np.nan, np.nan
else:
    print(" Insufficient data for VAR Forecast.")
    forecast_iv_skew, forecast_rv = np.nan, np.nan

# Step 7: Broadcast Forecast across dataset (for modularity / next-layer consumption)
block5_df['Forecast_IV_Skew'] = forecast_iv_skew
block5_df['Forecast_RV'] = forecast_rv

# Step 8: Theta Decay Calculation - Optimizing Time Decay Management
# Assuming ATM Theta decay is -0.05 per day per contract (approximate), modify based on actual market conditions.
theta_base = -0.05  
block5_df['Theta_Decay'] = theta_base * 100  # Adjusting for contract size (100 shares per contract)

# Step 9: Risk Management / Signal Logic - Blend Stress, Regime, VAR Forecast, and Theta Decay
high_stress_threshold = block5_df['Stress_Score'].quantile(0.80)

risk_condition = (
    (block5_df['Stress_Score'] > high_stress_threshold) &
    (block5_df['Regime_Label'] >= 4)
)

# Theta-Optimized Strategy Consideration
theta_condition = block5_df['Theta_Decay'] < -3  # If daily theta loss is significant (e.g., more than $3/day)

block5_df['Trade_Signal'] = np.where(
    risk_condition, 'Reduce Exposure / Hedge',
    np.where(theta_condition, 'Optimize for Theta Decay', 'Hold / Normal Trading')
)

# Step 10: Strategy Suggestion Layer - Based on Trade Signal & Theta Behavior
block5_df['Suggested_Strategy'] = np.where(
    block5_df['Trade_Signal'] == 'Reduce Exposure / Hedge',
    'Long Put / Protective Collar',
    np.where(block5_df['Trade_Signal'] == 'Optimize for Theta Decay',
             'Short Straddle / Short Strangle',  # Maximizing Theta Decay
             'Straddle / Strangle / Normal')  # Default if no stress/risk signal
)

# Step 11: Sample Output Preview - Now Fully Reframed
print("\n VRADS Block 5 - Final Signal Layer Sample Preview:")
print(block5_df[['Date', 'IV', 'IV_Skew', rv_column, 'Stress_Score', 'Regime_Label',
                 'Forecast_IV_Skew', 'Forecast_RV', 'Theta_Decay',
                 'Trade_Signal', 'Suggested_Strategy']].tail(5))

# Step 12: Save Block 5 Output for Advanced Research / Block 6
block5_df.to_csv('vac_block5_output_signal_layer.csv', index=False)

```
